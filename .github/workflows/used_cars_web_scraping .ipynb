```python
import selenium
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import os
import requests as rq
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager(). install ())
#driver=webdriver.Chrome(executable_path='C:\\driver\\newdrive\\chromedriver.exe')
```

```python
driver.get("https://www.mahindrafirstchoice.com/used-cars/Delhi")
```

```python
try:
    previous_height=driver.execute_script('return document.body.scrollHeight')
    while True:
        time.sleep(2)
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        time.sleep(2)
        new_height=driver.execute_script('return document.body.scrollHeight')
        if(new_height==previous_height):
            break
        previous_height=new_height
except Error as e:
    print (e)
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
##Price of the car
price1 = soup.find_all("span",{"class":"ft_price"})
pr1 = []
for _ in price1:
    pr1.append(_.text)



price = soup.find_all("span",{"class":"car_price"})
pr = []
for _ in price:
    pr.append(_.text)
    
pr2=[]
for ele in pr1:
    pr2.append(ele.strip())
    
n=len(pr2)
n
del pr[1:n+1]

########Location of the car
location = soup.find_all("span",{"class":"buyer_locations icon_set stock_location b"})
loc = []
for _ in location:
    loc.append(_.text)
    


    
```

```python
#######distance,fuel type, body type,owner

bottom_spec = soup.find_all("span",{"class":"overview_name"})
bot = []
for _ in bottom_spec:
    bot.append(_.text)
    
converted_list = []
for element in bot:
    converted_list.append(element.strip())
    
res = [ele.replace('\n', '') for ele in converted_list]


n=int((len(res)/4))
np_array = np.array(res)

reshaped_array = np.reshape(np_array, (n,4 ))
```

```python
data=pd.DataFrame(reshaped_array,columns=["Kilometers","Fuel","Body-type","owner"])
```

```python
data.head(10)
```

```python
#####year, make

top_spec = soup.find_all("h3",{"class":"buyer_make"})
top = []
for _ in top_spec:
    top.append(_.text)
    
split_list=[]
for ele in top:
    split_list.append(ele.split())
```

```python
df=pd.DataFrame(split_list,columns=["year","m","r"])
```

```python
df = df.replace([None], [''], regex=True)
```

```python
data['make'] = df['m'].str.cat(df['r'], sep =" ")
data["year"]=df["year"]
data["location"]=loc
data["Price"]=pr
```

```python
#####Model

top_spec1 = soup.find_all("h3",{"class":"buyer_variant"})
top1 = []
for _ in top_spec1:
    top1.append(_.text)
    
converted_list1 = []
for element in top1:
    converted_list1.append(element.strip())
    
model = [ele.replace('\n', '') for ele in converted_list1]


res = [ele.replace('\n', '') for ele in converted_list1]
```

```python
data["Model"]=model
```

```python
used_cars=pd.DataFrame(used_cars)
```

```python
used_cars=data[["year","make","Model","Price","location","Kilometers","Fuel","Body-type","owner"]]
```
